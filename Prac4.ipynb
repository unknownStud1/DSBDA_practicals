```python
# Preparing the data for training the model
import numpy as np

# np.c_[]: This is a numpy function used to concatenate arrays along the second axis (column-wise).

# LSTAT (Lower Status of the Population)
# RM (Average Number of Rooms per House)
x = pd.DataFrame(np.c_[dataset['LSTAT']  , dataset['RM']] , columns = ['LSTAT' , 'RM'])

# Assign the 'Price' column to the target variable y
y = dataset['MEDV']
```

```python
from sklearn.model_selection import train_test_split

''' x and y: These are the predictor variables (features) and the target variable, respectively.
test_size=0.2: This parameter specifies the proportion of the dataset that should be allocated to the testing set. 
Here, test_size=0.2 indicates that 20% of the data will be used for testing, while the remaining 80% will be used for training.
random_state=5: This parameter controls the randomness of the data splitting process.
Setting a fixed random_state ensures reproducibility, meaning that 
if you run the code multiple times with the same random_state, you'll get the same results.
It's common practice to set a random_state to maintain consistency in the data splits across different runs.

Output variables:
x_train and y_train: These variables store the training set of predictor variables and the corresponding target variable, 
respectively.
They will be used to train the machine learning model.
x_test and y_test: These variables store the testing set of predictor variables and the corresponding target variable, 
respectively. They will be used to evaluate the performance of the trained model on unseen data. ''' 


x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train , y_train)
```

```python
predictions = model.predict(x_test)

# Display the first 5 predictions
print("Predictions:")
print(predictions[:5])

# Display the corresponding actual values (ground truth) from the testing data
print("\nActual values:")
print(y_test[:5])
```
